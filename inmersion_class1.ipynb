# Tarea = Promedio de area de todos los inmuebles de los barrios // Graficar top 10

```python
df.Area.mean()
```

```python
Barrios = df.Barrio.unique()
Barrios
```

```python
barrio_promedio_area = {}
for barrio in Barrios:
    barrio_promedio_area[barrio] = df[df.Barrio == barrio].Area.mean()
barrio_promedio_area
```

```python
barrio_promedio_area = pd.Series(barrio_promedio_area).sort_values(ascending=False)
plt.figure(figsize=(20,10))
xvalues = barrio_promedio_area.head(10).index
yvalues = barrio_promedio_area.head(10).values
plt.plot(xvalues, yvalues)
plt.title('Promedio de area de los inmuebles por barrio')
plt.xlabel('Barrio')
plt.ylabel('Promedio de area')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x=barrio_promedio_area.head(10).index, y=barrio_promedio_area.head(10).values)
plt.title('Promedio de area de los inmuebles por barrio')
plt.xlabel('Barrio')
plt.ylabel('Promedio de area')
plt.xticks(rotation=90)
plt.show()
```
