```python
df.Precio = df.Precio.str.replace('.','',regex=True) # elimina puntos
```

```python
df['Precio_millon'] = df.Precio.astype(float)/1000000 # convierte a float y divide entre 1 millon
```

# Grafica de barras solo con plot

```python
df['Precio_millon'].plot(kind='hist', bins=50, figsize=(10,5))
```

# Grafica de barras con seaborn

```python
plt.figure(figsize=(10,6)) # define el tamaño de la gráfica 10 x 8
sns.histplot(df['Precio_millon'],
             kde=True ,
             bins=50,
             color='red',
             alpha=1,
             edgecolor='black',
             linewidth=1.2,
             stat='count',
             cumulative=False,
             common_norm=False,
             common_bins=False,
             discrete=False,
             element='bars',
             fill=True,
             line_kws=None,
             palette=None,
             log_scale=False,
             multiple='layer',
             shrink=1,
             legend=True,
             ax=None,
             )

plt.title('Histograma de precios',
          fontsize=24 ,
          fontweight='bold',
          color='red')
plt.xlabel('Precio',
           fontsize=18,
           fontweight='bold',
           color='blue')
plt.ylabel('Frecuencia',
           fontsize=18,
           fontweight='bold',
           color='yellow')
plt.xlim(50,2000)
plt.show()
```

# Grafica con Matplotlib

```python
plt.figure(figsize=(10,6),
           dpi=100,
           facecolor='grey',
           edgecolor='black',
           linewidth=2,
           frameon=True )
plt.hist(df['Precio_millon'],
         bins='auto',
         color='red',
         alpha=0.7,
         rwidth=0.85 ,
         edgecolor='black',
         linewidth=1.2 )
plt.title('Histograma de precios',
          fontsize=24 ,
          fontweight='bold',
          color='black')
plt.xlabel('Precio',
           fontsize=18,
           fontweight='bold',
           color='blue')
plt.ylabel('Frecuencia',
           fontsize=18,
           fontweight='bold',
           color='yellow')
plt.xlim(50,1000)
plt.savefig('C:\\Users\Omar\Desktop\INMERSION\histograma.jpeg')
plt.savefig('C:\\Users\Omar\Desktop\INMERSION\histograma.png')
plt.show()
```
